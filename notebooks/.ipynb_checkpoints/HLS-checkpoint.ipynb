```python
results = earthaccess.search_data(
    short_name="HLSS30",
    count=10
)
```

```python
for granule in results:
    # Get the metadata dictionary
    meta = granule['umm']
    
    # Cloud cover is typically in AdditionalAttributes
    for attr in meta.get('AdditionalAttributes', []):
        if attr['Name'] == 'CLOUD_COVERAGE':
            cloud_cover = attr['Values'][0]
            print(f"Granule: {granule['meta']['native-id']}")
            print(f"Cloud Cover: {cloud_cover}%")
            break
    
```
